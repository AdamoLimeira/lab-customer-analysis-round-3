```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
df = pd.read_csv('marketing_customer_analysis.csv')
df
```

```python
df.columns = df.columns.map(lambda x: x.replace(' ', '_'))
df
```

```python
sns.histplot(data = df['Response']).set_title('Total number of responses')
plt.show()
```

```python
df.Sales_Channel.unique()
```

```python
df_group = df.groupby('Sales_Channel')['Response'].value_counts(normalize = True).unstack()
df_group['Response_Rate']= df_group['Yes']*100
display(df_group.Response_Rate)
```

```python
df_group.reset_index(inplace = True) 
```

```python
plt.figure(figsize = (12,6))
sns.barplot(x = 'Sales_Channel', y = 'Response_Rate', data = df_group).set_title('Response % by Sales Channel')
plt.show()
```

```python
df_group2 = df.groupby('Total_Claim_Amount')['Response'].value_counts(normalize = True).unstack().fillna(0)
df_group2['Responsed_Rate'] = df_group2['Yes']*100
df_group2.reset_index(inplace = True)
df_group2
```

```python
my_bins = [*range(0,3500,500)] #* to unpack the results of range
print(my_bins)
df_group2['Claim_Amount_Bins']= pd.cut(df_group2['Total_Claim_Amount'], my_bins)
df_group2['Claim_Amount_Bins']
```

```python
plt.figure(figsize = (18,6))
sns.barplot(x = 'Claim_Amount_Bins', y = 'Responsed_Rate', data = df_group2)
plt.title('Response rate by intervals of total claim amounts')
plt.show()
```

```python
df_group3 = df.groupby('Income')['Response'].value_counts(normalize=True).unstack().fillna(0)
df_group3['Response_Rate'] = df_group3['Yes']*100
df_group3.reset_index(inplace= True)
```

```python
my_bins2 = [0, 30000, 50000, 75000, 100000]
df_group3['Income_Bins']= pd.cut(df_group3['Income'], my_bins2)
df_group3['Income_Bins']
```

```python
plt.figure(figsize = (16,8))
sns.barplot(x = 'income_bins', y = 'response_rate', data = df_group3)
plt.title('Response Rate by income intervals')
plt.show()
```
